```python
#1.1 Carga el dataset en un DataFrame de Pandas

import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/Bootcamp/migracion.csv')
df
```

```python
#1.2 Identifica y trata valores perdidos en el dataset
#revisamos si algun valor nulo por columna

df.isna().sum()

#no hay! :D
```

```python
# 1.3 Detecta y filtra outliers usando el método del rango intercuartílico (IQR).

#calculamos el rango

q1 = df['Cantidad_Migrantes'].quantile(0.25)
q3 = df['Cantidad_Migrantes'].quantile(0.75)
IQR = q3 - q1

#establecemos los limites
lim_inf = q1 - 1.5 * IQR
lim_sup = q3 + 1.5 * IQR

#los imprimo para revisarlos
print(lim_inf)
print(lim_sup)
```

```python
# 1.3 filtramos el dataframe para sacar los outliers

outliers = df[(df['Cantidad_Migrantes'] < lim_inf) |(df['Cantidad_Migrantes'] > lim_sup)]

outliers

#no hay outliers para la columna Cantidad_Migrantes :D
```

```python
#Reemplaza los valores de la columna "Razon_Migracion" usando mapeo de valores
#(ejemplo: "Económica" → "Trabajo", "Conflicto" → "Guerra").

razones = {'Económica':'Trabajo', 'Conflicto':'Guerra', 'Educativa': 'Educacion'}

df['Razon_Migracion'] = df['Razon_Migracion'].map(razones)

df
```

```python
#2.3 #Usa .value_counts() para contar cuántos movimientos de migración ocurrieron por cada razón

df['Razon_Migracion'].value_counts()
```

```python
#3.1 Agrupa los datos por "Razon_Migracion" y calcula la suma total de migrantes para cada categoría

df.groupby('Razon_Migracion')['Cantidad_Migrantes'].sum()
```

```python
#3.2 Obtén la media del IDH de los países de origen por cada tipo de migración

df.groupby('Razon_Migracion')['IDH_Origen'].mean()
```

```python
#3.3 Ordena el DataFrame de mayor a menor cantidad de migrantes.

df = df.sort_values('Cantidad_Migrantes', ascending = False)

df
```

```python
#4.1 Filtra y muestra solo las migraciones por conflicto

df_conflicto = df[df['Razon_Migracion'] == 'Guerra']

df_conflicto
```

```python
#4.2 Selecciona y muestra las filas donde el IDH del país de destino sea mayor a 0.90

df_idh = df[df['IDH_Destino'] > 0.9]

df_idh
```

```python
#4.3 Crea una nueva columna "Diferencia_IDH" que calcule la diferencia de IDH entre país de origen y destino

df['Diferencia_IDH'] = df['IDH_Origen'] - df['IDH_Destino']

df
```

```python
#Guarda el DataFrame final en un nuevo archivo CSV llamado "Migracion_Limpio.csv", sin el índice

df.to_csv('/content/drive/My Drive/Bootcamp/Migracion_Limpio.csv', index=False)
```
