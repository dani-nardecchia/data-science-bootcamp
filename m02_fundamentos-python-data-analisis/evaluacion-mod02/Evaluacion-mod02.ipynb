```python
#6.Usa un diccionario para almacenar descuentos especiales por autor.

descuentos = {'Han Kang': 0.2, 'Alejandro Dumas': 0.4, 'Agustina Bazterrica': 0, 'Nnedi Okorafor': 0, 'Neil Postman': 10}
```

```python
#Ahora modificamos la funcion para agregar los descuentos
def comprar_libros(titulo: str, cantidad: int):
  '''

  '''
  # Verificar si el libro está en el inventario y si la cantidad deseada está disponible
  for libro in libros:
    if titulo == libro['Título']:
      if cantidad <= libro['Stock']:
        libro['Stock'] = libro['Stock'] - cantidad   #con esto updateamos el stock del libro comprado
        monto_compra = libro['Precio'] * cantidad    #mostramos el monto total de la compra

        descuento = 0                            #ya obtuvimos el precio ahora el descuento
        autor = libro['Autor']
        for key in descuentos.keys():
          if key == autor:
            descuento = descuentos[key]
            break

        descuento_aplicado = monto_compra * descuento
        precio_final = round((monto_compra - descuento_aplicado),2)
        return precio_final, descuento_aplicado, cantidad

      elif cantidad > libro['Stock']:
        print('No se encuentra la cantidad solicitada')
```

```python
#corremos la funcion de nuevo, esta vez con los descuentos aplicados

titulo = input('Ingrese el título que desea comprar:')
cantidad = int(input('Ingrese la cantidad que desea comprar:'))

monto_compra,descuento, cantidad  = comprar_libros(titulo, cantidad)  #monto compra me guarda lo que vale cada compra
total_compras = monto_compra
total_descuento = descuento

resp_client = input('Desea hacer otra compra? Presione S para Si o N para No')

while resp_client == 'S':
  titulo = input('Ingrese el título que desea comprar:')
  cantidad = int(input('Ingrese la cantidad que desea comprar:'))

  monto_compra, descuento, cantidad = comprar_libros(titulo, cantidad)
  total_compras = total_compras + monto_compra
  total_descuento = total_descuento + descuento

  resp_client =input('Desea hacer otra compra? Presione S para Si o N para No')
  if resp_client == 'N':
    print(f'El total de su compra es: {total_compras}')
    break
```

```python
# 7. Al finalizar la compra, muestra un resumen con el total de libros comprados, el monto total pagado y el ahorro por descuentos.
#Para poder mostrar esta informacion, necesitamos añadir estas cantidades como variables a nuestro ciclo while

total_compras = 0 #sera el total acumulado dsps de varias compras y la variable q usaremos en la factura final
cant_compra = 0 #sera la cantidad de libros comprados, acumulandose por cada ciclo
total_descuento = 0  #ahorro acumulable por cada compra

titulo = input('Ingrese el título que desea comprar:')
cantidad = int(input('Ingrese la cantidad que desea comprar:'))

monto_compra, descuento, cantidad = comprar_libros(titulo, cantidad)  #monto compra me guarda lo que vale cada compra

total_compras = monto_compra
cant_compra = cantidad
total_descuento = descuento

resp_client = input('Desea hacer otra compra? Presione S para Si o N para No')

while resp_client == 'S':
  titulo = input('Ingrese el título que desea comprar:')
  cantidad = int(input('Ingrese la cantidad que desea comprar:'))

  monto_compra, descuento, cantidad = comprar_libros(titulo, cantidad)

  total_compras = total_compras + monto_compra
  total_descuento = total_descuento + descuento
  cant_compra = cant_compra + cantidad

  resp_client =input('Desea hacer otra compra? Presione S para Si o N para No')
  if resp_client == 'N':
    total_compras = round(total_compras,2)
    print(f"""
-----------------------------
          FACTURA
-----------------------------
Total a pagar: $ {total_compras:.2f}
Items comprados: {cant_compra}
Su ahorro fue de: $ {total_descuento:.2f}
Gracias por su compra.""")
    break
```
